```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading datasets

df1 = pd.read_csv('Downloads/eruptions.csv')
df2 = pd.read_csv('Downloads/volcano.csv')
```

```python
df1.head(10)
```

```python
df2.head(10)
```

```python
df2['evidence_category']
```

```python
#merging datasets (inner join)

df = pd.merge(df1, df2, on='volcano_number', how='inner')
df
```

```python
df.info()
```

```python
#not using df_unique because unique entries per volcano doesn't include all eruptions for each volcano along the years

#df_unique = pd.merge(df1, df2, right_index=True, left_index=True)
#df_unique.head()
```

```python
#dropping duplicate features and renaming the remainig ones

df = df.drop(columns=['volcano_name_y', 'latitude_y', 'longitude_y'])
df.rename(columns={'volcano_name_x': 'volcano_name','latitude_x':'latitude', 'longitude_x':'longitude'}, inplace=True)
```

```python
#filtering dataset to get only significant eruptions (VEI > 5)
#decide whether to filter significant eruptions or not, given it will drop sample size to 50 entries

df_new = df[df['vei'] > 5] 

df_new.shape
```

```python
#data is imbalanced (mostly VEI 6 than 7, no 8), small sample (50 entries)

df_new.describe()
```

```python
#correlation matrix

#my_cmap = sns.light_palette('purple', as_cmap=True)

plt.figure(figsize=(14,8))
sns.heatmap(df_new.corr(),annot=True, cmap='RdPu') 
plt.show()
```

```python
#creating subset of df to create scatter plot matrix

#df_sub = df_new.drop(columns=['volcano_number', 'eruption_number',  'latitude', 'longitude'])
df_sub = df_new.drop(columns=['volcano_number', 'eruption_number', 'start_year', 'start_month', 'start_day','end_year', 'end_month', 'end_day', 'latitude', 'longitude'])
```

```python
#scatter plot matrix
#verify if its best to use the population mean because of the data repetition on datasets migration

sns.set_theme(style='ticks')
sns.pairplot(df_sub, hue='vei')
```

```python
#boxplot of elevation by volcano type

plt.figure(figsize=(10,6))
sns.set_style('darkgrid', {'axes.facecolor': '.9'})

g = sns.boxplot(x=df_new.elevation, y=df_new.primary_volcano_type, palette='magma')
g = sns.swarmplot(x=df_new.elevation, y=df_new.primary_volcano_type, color='.45')

plt.xlabel("Elevation")
plt.ylabel("")
plt.title("Elevation by Volcano Type")

plt.show()
```

```python
#boxplot of elevation by tectonic settings

plt.figure(figsize=(10,6))
sns.set_style('darkgrid', {'axes.facecolor': '.9'})

g = sns.boxplot(x=df_new.elevation, y=df_new.tectonic_settings, palette='magma')
g = sns.swarmplot(x=df_new.elevation, y=df_new.tectonic_settings, color='.45')

plt.xlabel("Elevation")
plt.ylabel("")
plt.title("Elevation by Tectonic Settings")

plt.show()
```

```python
#bar plot of significant volcano by country

fig, ax = plt.subplots(figsize=(18,12))
sns.set_style('darkgrid', {'axes.facecolor': '.9'})

sns.countplot(data=df_new, y='country', hue='vei', ax=ax, palette='magma')

plt.xlabel("Volcano Count")
plt.ylabel("Country")
plt.title("Significant Volcanoes by Country")

plt.show()
```

```python
#bar plot of significant volcano by region

fig, ax = plt.subplots(figsize=(18,12))
sns.set_style('darkgrid', {'axes.facecolor': '.9'})

sns.countplot(data=df_new, y='region', hue='vei', ax=ax, palette='magma')

plt.xlabel("Volcano Count")
plt.ylabel("Region")
plt.title("Significant Volcanoes by Region")

plt.show()
```

```python
#bar plot of eruptions by volcano
#needs formatting

fig, ax = plt.subplots(figsize=(16,8))
#sns.set_style('darkgrid', {'axes.facecolor': '.9'})

df_new.groupby(['volcano_name']).count()['eruption_number'].plot.bar(ax=ax)

plt.xlabel("Volcano Name")
plt.ylabel("Eruptions Count")
plt.title("Eruptions by Volcano")

plt.show()
```
