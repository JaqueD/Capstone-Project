```python
fig = df_new[df_new.VEI==6].plot(kind='scatter',x='Total_Deaths',y='Elevation',color='orange', label='VEI 6')
df_new[df_new.VEI==7].plot(kind='scatter',x='Total_Deaths',y='Elevation',color='blue', label='VEI 7',ax=fig)
df_new[df_new.VEI==8].plot(kind='scatter',x='Total_Deaths',y='Elevation',color='green', label='VEI 8', ax=fig)
fig.set_xlabel("Total Deaths")
fig.set_ylabel("Elevation")
fig.set_title("Total Deaths by Elevation")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```
